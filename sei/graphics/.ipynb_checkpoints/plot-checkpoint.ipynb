```python
import os

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

<br>

# SNPs distribution

> Load data

```python
def load_data(model):
    # Path data
    path_data = "../../Data/Msprime/snp_distribution/sfs_{}/".format(model)
    
    data = pd.read_json(path_or_buf="{}SFS_{}-all.json".format(path_data, model), typ='frame')
    
    return data
```

```python
sfs_decline = load_data("decline")
```

```python
sfs_decline
```

<br>

```python
# Compute log10 of kappa & tau
sfs_decline['Log10(Tau)'] = sfs_decline['Parameters'].apply(lambda ele: np.log10(ele['Tau']))
sfs_decline['Log10(Kappa)'] = sfs_decline['Parameters'].apply(lambda ele: np.log10(ele['Kappa']))

# Compute mean of SNPs (then log10)
sfs_decline['SNPs'] = sfs_decline['SNPs'].apply(lambda ele: np.mean(ele))
somme = sum(sfs_decline['SNPs'])
sfs_decline['SNPs'] = sfs_decline['SNPs'].apply(lambda ele: ele/somme)
```

```python
sfs_bis = sfs_decline[['Log10(Tau)', 'Log10(Kappa)', 'SNPs']]
sfs_bis
```

```python
sfs_bis = sfs_bis.groupby(['Log10(Tau)', 'Log10(Kappa)']).mean().unstack('Log10(Kappa)')
sfs_bis
```

> Plot data

```python
def snp_distribution_3d(data):
    # Plot
    sns.set_theme(style='whitegrid')

    # Plot
    sns.heatmap(data)
        
    plt.show()
```

```python
%matplotlib widget
snp_distribution_3d(sfs_bis)
```
