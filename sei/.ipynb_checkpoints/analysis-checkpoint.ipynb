# SFS Shape

Method to check the SFS obtained with msprime.

1. The SFS of a **constant population** fits well to the theoretical SFS of any constant population

2. The SFS of an **decline model**

> tau time ago the population decreased from size Ne*kappa to size Ne

3. The SFS of an **growth model**

> tau time ago the population increased from size Ne to size Ne*kappa

4. The SFS of a **migration model**

> Population 1 of size pop1 and population 2 of size pop2 = kappa*pop1. There are some migrations from population 2 to 1 with m12

<br>

### Generate Data

```python
def simulation_parameters(sample, ne, rcb_rate, mu, length):
    """
    Set up the parametres for the simulation with msprime.
    """
    parameters = {
        "sample_size": sample, "size_population": ne, "rcb_rate": rcb_rate, "mu": mu,
        "length": length
    }
    return parameters

def generate_sfs():
    """
    Generate the SFS of various demographic scenario:
      
      - Constant population
      - Theoritical SFS of any constant population
      - Sudden decline scenario
      - Sudden growth scenario
      - Migration 
    """
    # Parameters for the simulation
    params = simulation_parameters(sample=10, ne=1, rcb_rate=2e-2, mu=2e-2, length=1e5)

    # Constant scenario
    sfs_cst = ms.msprime_simulation(model=ms.constant_model, params=params, debug=False)

    # Theoretical SFS for any constant population
    sfs_theorique = [0] * (params["sample_size"] - 1)
    for i in range(len(sfs_theorique)):
        sfs_theorique[i] = 1 / (i+1)
    
    # Define tau & kappa for decline/growth scenario
    params.update({"Tau": 1.0, "Kappa": 10.0})

    sfs_decline = ms.msprime_simulation(model=ms.sudden_decline_model, params=params, debug=False)

    sfs_growth = ms.msprime_simulation(model=ms.sudden_growth_model, params=params, debug=False)

    # Migration scenario
    # Define Kappa, m12 and m21
    params.update({"Kappa": 10.0, "m12": 1.0, "m21": 0})
    
    sfs_migration = ms.msprime_simulation(model=ms.two_pops_migration_model, params=params, debug=False)
    
    return sfs_cst, sfs_theorique, sfs_decline, sfs_growth, sfs_migration
```

```python
sfs = generate_sfs()
```

<br>

### Plot

```python
def plot_sfs(sfs, label, color, title, style, axis=False):
    """
    Graphic representation of Site Frequency Spectrum (SFS)

    Parameter
    ---------
    sfs: list
        list of sfs to plot
    label: list
        the label of each sfs
    color: list
        the color of each curve
    style: list
        the linestyle
    title: string
        title of the plot
    """
    # Plot
    plt.figure(figsize=(16,12))
    for i, spectrum in enumerate(sfs):
        normalized_spectrum = [ele / sum(spectrum) for ele in spectrum]  # normalization of each sfs
        plt.plot(normalized_spectrum, color=color[i], linestyle=style[i], label=label[i])

    # Caption
    plt.legend(loc="upper right", fontsize="large")

    # Label axis
    plt.xlabel("Allele frequency", fontsize="large")
    plt.ylabel("Percent of SNPs", fontsize="large")

    # X axis values
    if axis:
        x_ax, x_values = [], []
        for i in range(len(sfs[0])):
            x_ax.append(i)
            x_values.append("{}/{}".format(i+1, len(sfs[0])+1))
        plt.xticks(x_ax, x_values)

    # Title + save plot to the folder ./Figures
    plt.title(title, fontsize="xx-large")
    plt.show()


plot_sfs(
    sfs=sfs,
    label=["Constant", "Theoretical", "Declin", "Growth", "Migration"],
    color=["tab:blue", "tab:orange", "tab:red", "tab:green", "tab:gray"],
    style=["solid", "dashed", "solid", "solid", "solid"],
    title="Unfold SFS for various scenarios", axis=True
)
```
